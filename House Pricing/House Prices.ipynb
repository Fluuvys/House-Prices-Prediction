```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import dataset
train_df = pd.read_csv("./data/train.csv")
test_df = pd.read_csv("./data/test.csv")
test_ids = test_df["Id"]
```

#### Handle missing data

```python
#Identify missing data
missing_data = train_df.isna().sum()
missing_columns = missing_data[missing_data>0]
print(missing_columns)
```

```python
from sklearn.impute import SimpleImputer
# I will drop columns with too many missing values since they might not contribute much to the model
columns_to_drop = ['Alley', 'PoolQC', 'MiscFeature']
train_df = train_df.drop(columns_to_drop, axis = 1)

#Imputing missing values with the mean for numerical columns
num_cols = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
imputer = SimpleImputer(strategy='mean')
train_df[num_cols] = imputer.fit_transform(train_df[num_cols])

#Imputing missing values with mode for categorical features
cat_cols = ['MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Electrical', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'Fence']
for col in cat_cols:
    train_df[col].fillna(train_df[col].mode()[0], inplace = True)
```

```python
train_df.head()
```

```python
missing_data = train_df.isna().sum()
missing_columns = missing_data[missing_data>0]
if missing_data[missing_data > 0].empty:
    print("All missing data has been handled!")
else:
    print("There are still some missing values:")
    print(missing_data[missing_data > 0])
```

#### Handle Categorical Data

```python
categorical_columns = train_df.select_dtypes(include = ['object', 'category', 'bool']).columns.to_list()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for col in categorical_columns:
    train_df[col] = label_encoder.fit_transform(train_df[col])
```

```python
train_df.head()
```

#### Split the data into training and testing data and train the model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

y = train_df["SalePrice"]
X = train_df.drop("SalePrice", axis = 1)
#Need to scale down data in order to train model more effectiveley.
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)


X_train, X_val, y_train,  y_val = train_test_split(X_scaled,y, test_size = 0.25, random_state = 42)
```

```python
model = RandomForestRegressor().fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
predictions = model.predict(X_val)
mse = mean_squared_error(y_val, predictions)
r2 = r2_score(y_val, predictions)
print("MSE:", mse)
print("R-squared:", r2)
```

```python
#Identify missing data
missing_data = test_df.isna().sum()
missing_columns = missing_data[missing_data>0]
print(missing_columns)
```

```python
from sklearn.impute import SimpleImputer

# Drop columns with excessive missing values
columns_to_drop = ['Alley', 'PoolQC', 'MiscFeature']
test_df = test_df.drop(columns_to_drop, axis=1)

# Imputing missing values for numerical columns with median
num_cols = ['LotFrontage', 'MasVnrArea', 'GarageYrBlt', 'BsmtFinSF1', 
            'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'BsmtFullBath', 
            'BsmtHalfBath', 'GarageCars', 'GarageArea']
imputer = SimpleImputer(strategy='median')
test_df[num_cols] = imputer.fit_transform(test_df[num_cols].copy())

# Imputing missing values for categorical columns with mode
cat_cols = ['MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 
            'BsmtFinType1', 'BsmtFinType2', 'FireplaceQu', 'GarageType', 
            'GarageFinish', 'GarageQual', 'GarageCond', 'Fence', 
            'MSZoning', 'Utilities', 'Exterior1st', 'Exterior2nd', 
            'KitchenQual', 'Functional', 'SaleType']

for col in cat_cols:
    test_df[col].fillna(test_df[col].mode()[0], inplace=True)
```
